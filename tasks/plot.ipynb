```python
from tqdm import tqdm_notebook as tqdm
import pandas as pd
import json
import itertools
import pickle
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
df_fda = pd.read_csv("./data/clue_to_fda.csv", sep =",", header=None)
df_info = pd.read_csv("./data/drug_info_2.0.csv", sep =",")

with open("./data/99_our_v0.6_py2.pkl", 'rb') as f:
    df_embed = pickle.load(f)
```

```python
df_fda = df_fda[df_fda[3] == 'fda']
fda_list = df_fda[1].unique()
```

```python
df_info
```

```python
dict_fda_embed = {}
dict_fda_moa = {}
for fda in fda_list:
    moa = df_info[df_info['pert_id'] == fda]['moa'].values[0].split("|")[0]
    embed = df_embed[fda][0]
    
    dict_fda_embed[fda] = embed
    dict_fda_moa[fda] = moa
```
